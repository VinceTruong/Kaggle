```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv")

df.head(10)
```

```python
df.shape
```

**There are 891 sample for us to analyse**

**Because PassengerId, Name and Ticket cannot convert to real number and has minor impact on the result, then I remove them**

```python
df.drop(['PassengerId','Name', 'Ticket'], axis=1, inplace=True)

df.head(10)
```

**Now I convert Sex into binary number**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE = LabelEncoder()
```

```python
df['Sex'] = LE.fit_transform(df['Sex'])

df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Cabin'].unique()
```

**There are many type of cabin and with my basic competence, I will drop this column**

```python
df.drop('Cabin', axis=1, inplace=True)

df.head()
```

**Embarked column has 2 missing values. Then, I fill the empty entries with the most frequency value in Embarked**

```python
df['Embarked'].value_counts()
```

```python
df['Embarked'].fillna('S', inplace=True)

df.head()
```

**I convert Embraked columns into 3 other columns by using get_dummies function()**

```python
dummy = pd.get_dummies(df['Embarked'])

dummy.head(10)
```

**And I drop the last column**

```python
dummy.drop('S', axis=1, inplace=True)

dummy
```

```python
new_df = pd.concat([df, dummy], axis=1)

new_df.head(10)
```

```python
new_df.info()
```

**Column 'Age' remain empty entries. I fill mean of age into these entries**

```python
from math import *

age_mean = floor(np.mean(new_df['Age']))

new_df['Age'].fillna(age_mean, inplace=True)

new_df.head(10)
```

```python
new_df.drop('Embarked', axis=1, inplace=True)

new_df.head()
```

**We look the desciption again**

```python
new_df.describe()
```

**No missing value!**

```python
X = new_df.drop('Survived', axis=1)
X.head()
```

```python
y = new_df['Survived']
y.head()
```

# Build Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import cross_val_score
```

**LogisticRegression**

```python
cross_val_score(LogisticRegression(solver='liblinear',multi_class='ovr'), X, y, cv=5).mean()
```

**SVC**

```python
cross_val_score(SVC(gamma='auto'), X, y, cv=5).mean()
```

**RandomForestClassifier**

```python
cross_val_score(RandomForestClassifier(n_estimators=70), X, y, cv=5).mean()
```

**DecisionTreeClassifier**

```python
cross_val_score(DecisionTreeClassifier(), X, y, cv=5).mean()
```

## Then we use RandomForestClassifier model

```python
RFC = RandomForestClassifier()
```

```python
RFC.fit(X, y)
```
